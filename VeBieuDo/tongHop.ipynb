```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Äá»c dá»¯ liá»u
data = pd.read_csv(r"D:\hocTap\CODE\Python\doAn\weather_classification_data.csv")

# HÃ m Äá» thÃªm nhÃ£n giÃ¡ trá» trÃªn cá»t
def add_value_labels(ax):
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.1f'), 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 8), textcoords='offset points')
def add_value_labels(ax):
    for line in ax.lines:  # Xá»­ lÃ½ tá»«ng ÄÆ°á»ng trong biá»u Äá»
        for x, y in zip(line.get_xdata(), line.get_ydata()):
            ax.annotate(f'{y:.2f}',  # In giÃ¡ trá» vá»i 2 chá»¯ sá» tháº­p phÃ¢n
                        (x, y), 
                        textcoords="offset points",
                        xytext=(0, 5),  # Dá»ch chuyá»n nhÃ£n lÃªn trÃªn
                        ha='center', 
                        fontsize=10, 
                        color='black')

# Váº½ Bar Plot cho nhiá»t Äá» theo khu vá»±c
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x='Location', y='Temperature', errorbar=None, hue='Location', legend=False)
plt.title('Nhiá»t Äá» Trung bÃ¬nh giá»¯a cÃ¡c Khu vá»±c', fontsize=16)
plt.xlabel('Khu vá»±c', fontsize=12)
plt.ylabel('Nhiá»t Äá» Trung bÃ¬nh (Â°C)', fontsize=12)
plt.grid(axis='y')
add_value_labels(ax)
plt.show()

# Váº½ Bar Plot cho tá»c Äá» giÃ³ theo khu vá»±c
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x='Location', y='Wind Speed', errorbar=None, hue='Location', legend=False)
plt.title('Tá»c Äá» GiÃ³ Trung bÃ¬nh giá»¯a cÃ¡c Khu vá»±c', fontsize=16)
plt.xlabel('Khu vá»±c', fontsize=12)
plt.ylabel('Tá»c Äá» GiÃ³ Trung bÃ¬nh (km/h)', fontsize=12)
plt.grid(axis='y')
add_value_labels(ax)
plt.show()

# Váº½ Bar Plot cho lÆ°á»£ng mÆ°a theo khu vá»±c
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x='Location', y='Precipitation (%)', errorbar=None, hue='Location', legend=False)
plt.title('LÆ°á»£ng MÆ°a Trung bÃ¬nh giá»¯a cÃ¡c Khu vá»±c', fontsize=16)
plt.xlabel('Khu vá»±c', fontsize=12)
plt.ylabel('LÆ°á»£ng MÆ°a Trung bÃ¬nh (%)', fontsize=12)
plt.grid(axis='y')
add_value_labels(ax)
plt.show()

# Váº½ Bar Plot cho Äá» áº©m theo khu vá»±c
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x='Location', y='Humidity', errorbar=None, hue='Location', legend=False)
plt.title('Äá» áº©m Trung bÃ¬nh giá»¯a cÃ¡c Khu vá»±c', fontsize=16)
plt.xlabel('Khu vá»±c', fontsize=12)
plt.ylabel('Äá» áº©m Trung bÃ¬nh (%)', fontsize=12)
plt.grid(axis='y')
add_value_labels(ax)
plt.show()

uv_location = data.groupby('Location')['UV Index'].mean().reset_index()
# Váº½ biá»u Äá» Line Plot cho UV theo khu vá»±c
plt.figure(figsize=(8, 6))
ax = sns.lineplot(data=uv_location, x='Location', y='UV Index', marker='o', color='orange', linewidth=2.5)
plt.title('Biáº¿n Äá»ng chá» sá» UV giá»¯a cÃ¡c khu vá»±c', fontsize=16)
plt.xlabel('Khu vá»±c', fontsize=12)
plt.ylabel('Chá» sá» UV trung bÃ¬nh', fontsize=12)
plt.grid(True)
add_value_labels(ax)
plt.tight_layout()  # Äiá»u chá»nh khÃ´ng gian giá»¯a cÃ¡c nhÃ£n
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv(r"D:\hocTap\CODE\Python\doAn\weather_classification_data.csv")
# Lá»c cÃ¡c cá»t sá» trong dá»¯ liá»u
numerical_columns = data.select_dtypes(include=['float64', 'int64'])

# TÃ­nh ma tráº­n tÆ°Æ¡ng quan
correlation_matrix = numerical_columns.corr()

# Váº½ Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Heatmap biá»u diá»n má»i liÃªn há» tÆ°Æ¡ng quan giá»¯a cÃ¡c dá»¯ liá»u", fontsize=16)
plt.show()
```
